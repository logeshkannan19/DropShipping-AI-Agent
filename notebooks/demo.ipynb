## 2. Demand Prediction

Predict demand for a product.

```python
from backend.models.demand_prediction import DemandPredictor

predictor = DemandPredictor()

features = {
    'category': 'Electronics',
    'price': 35.99,
    'rating': 4.5,
    'review_count': 200,
    'competitor_count': 25,
    'supplier_rating': 4.7,
    'shipping_days': 12
}

result = predictor.predict(features)
print(f"Demand Score: {result['demand_score']:.2%}")
print(f"Confidence: {result['confidence']:.2%}")
```

## 3. Pricing Optimization

Optimize pricing for a product.

```python
from backend.agents.pricing_optimization import PricingOptimizationAgent

pricing = PricingOptimizationAgent(target_margin=30)

result = pricing.optimize_price(
    cost_price=25.99,
    competitor_prices=[45.99, 49.99, 42.99],
    demand_score=0.75
)

print(f"Optimal Price: ${result['optimal_price']:.2f}")
print(f"Margin: {result['margin_percent']:.1f}%")
print(f"Strategy: {result['strategy']}")
```
